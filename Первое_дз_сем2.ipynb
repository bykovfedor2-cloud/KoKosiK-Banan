Задача 1: Как, по словам деда, он добирался каждое утро в школу

```python
from heapq import heappush, heappop

def your_code(n, m, s, t, edges):

    g = [[] for _ in range(n)]
    for u, v, w in edges:
        g[u].append((v, w))


    INF = 10**18
    dist = [INF] * n
    dist[s] = 0
    prev = [-1] * n


    q = []
    heappush(q, (0, s))

    while q:
        d, v = heappop(q)
        
        if d > dist[v]:
            continue
            
        for to, weight in g[v]:
            if dist[to] > dist[v] + weight:
                dist[to] = dist[v] + weight
                prev[to] = v
                heappush(q, (dist[to], to))

    if dist[t] == INF:
        return "не могу найти машрут"

    path = []
    current = t
    while current != -1:
        path.append(current)
        current = prev[current]

    path.reverse()
    
    return f"{dist[t]}\n{' '.join(map(str, path))}"


# ────────────────────────────────────────────────
# ТЕСТЫ
# ────────────────────────────────────────────────

def test_gps_cases():
    tests = [
        # Тест 1:
        {
            "n": 8, "m": 8, "s": 0, "t": 7,
            "edges": [(0,1,600),(0,2,1200),(1,3,800),(2,3,400),(2,4,700),(3,5,500),(4,5,300),(5,7,900)],
            "answer": "2800\n0 1 3 5 7"
        },
        # Тест 2:
        {
            "n": 4, "m": 3, "s": 0, "t": 3,
            "edges": [(0,1,1),(1,2,1),(0,2,10)],
            "answer": "Ne mogu naiti marshrut! :'("
        },
        # Тест 3:
        {
            "n": 4, "m": 4, "s": 0, "t": 3,
            "edges": [(0,1,4),(0,3,12),(1,2,3),(2,3,5)],
            "answer": "12\n0 3"
        },
        # Тест 4:
        {
            "n": 1, "m": 0, "s": 0, "t": 0,
            "edges": [],
            "answer": "0\n0"
        },
        # Тест 5:
        {
            "n": 3, "m": 3, "s": 0, "t": 2,
            "edges": [(0,1,5),(0,2,12),(1,2,3)],
            "answer": "8\n0 1 2"
        }
    ]

    for i, test in enumerate(tests, 1):
        print(f"Тест {i}:")
        result = your_code(test["n"], test["m"], test["s"], test["t"], test["edges"])
        print(f"Ожидаемый: {test['answer']}")
        print(f"Получено:  {result}")
        print("-" * 50)


if __name__ == "__main__":
    test_gps_cases()
```


Задача 2: Дорога в Шереметьево

```python
def your_code(n, tariffs):
    INF = 10**9
    dist = [row[:] for row in tariffs]  

    for i in range(n):
        for j in range(n):
            if dist[i][j] == -1:
                dist[i][j] = INF

    for k in range(n):
        for i in range(n):
            for j in range(n):
                if dist[i][k] < INF and dist[k][j] < INF:
                    dist[i][j] = min(dist[i][j], dist[i][k] + dist[k][j])

    for i in range(n):
        for j in range(n):
            if dist[i][j] >= INF:
                dist[i][j] = 999999
    
    return dist


def print_matrix(matrix):
    for row in matrix:
        print(' '.join(str(x) if x < 999999 else '-1' for x in row))


test_cases = [
    # Тест 1:
    { 
        "n": 5, 
        "tariffs": [
            [0, 10, -1, -1, 25], 
            [-1, 0, 15, 5, -1], 
            [-1, -1, 0, 8, 12], 
            [20, -1, -1, 0, 10], 
            [-1, 30, -1, -1, 0]
        ], 
        "expected": [
            [0, 10, 25, 15, 25], 
            [25, 0, 15, 5, 15], 
            [28, 38, 0, 8, 12], 
            [20, 30, 45, 0, 10], 
            [55, 30, 45, 35, 0]
        ]
    },
    # Тест 2: Полносвязный
    { 
        "n": 3, 
        "tariffs": [[0,5,9],[16,0,12],[7,3,0]], 
        "expected": [[0,5,9],[16,0,12],[7,3,0]] 
    },
    # Тест 3: Недостижимые города
    { 
        "n": 4, 
        "tariffs": [
            [0,10,-1,-1],
            [-1,0,20,-1],
            [-1,-1,0,30],
            [-1,-1,-1,0]
        ], 
        "expected": [
            [0,10,30,60],
            [999999,0,20,50],
            [999999,999999,0,30],
            [999999,999999,999999,0]
        ] 
    }
]


# Запуск тестов
if __name__ == "__main__":
    for i, test in enumerate(test_cases, 1):
        print(f"Тест {i}")
        print(f"n = {test['n']}")
        
        result = your_code(test['n'], test['tariffs'])
        
        print("Результат:")
        print_matrix(result)
        
        print("\nОжидаемый:")
        for row in test['expected']:
            print(' '.join(str(x) if x < 999999 else '-1' for x in row))
        
        print("-" * 60)
```
