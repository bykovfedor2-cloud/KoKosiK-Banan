Периодика

```python
def lps_make(s):
    n=len(s)
    a=[0]*n
    j=0
    i=1
    while i<n:
        if s[i]==s[j]:
            j+=1
            a[i]=j
            i+=1
        else:
            if j>0:
                j=a[j-1]
            else:
                a[i]=0
                i+=1
    return a

def per(s):
    n=len(s)
    if n==0:
        return 0
    p=lps_make(s)
    k=n-p[-1]
    if n%k==0:
        return k
    return n

def solve1():
    t=int(input())
    for _ in range(t):
        s=input().strip()
        print(per(s))
        if _!=t-1:
            print()

solve1()
```

Манакер

```python
def pal(s):
    n=len(s)
    if n==0:
        return 0

    d1=[0]*n
    l=0
    r=-1
    i=0
    while i<n:
        if i>r:
            k=1
        else:
            k=min(d1[l+r-i], r-i+1)
        while i-k>=0 and i+k<n and s[i-k]==s[i+k]:
            k+=1
        d1[i]=k
        if i+k-1>r:
            l=i-k+1
            r=i+k-1
        i+=1

    d2=[0]*n
    l=0
    r=-1
    i=0
    while i<n:
        if i>r:
            k=0
        else:
            k=min(d2[l+r-i+1], r-i+1)
        while i-k-1>=0 and i+k<n and s[i-k-1]==s[i+k]:
            k+=1
        d2[i]=k
        if i+k-1>r:
            l=i-k
            r=i+k-1
        i+=1

    return sum(d1)+sum(d2)

def solve2():
    s=input().strip()
    print(pal(s))

solve2()
```

Циклики

```python
def sufarr(s):
    s=s+'$'
    n=len(s)
    p=[0]*n
    c=[0]*n
    a=[]
    i=0
    while i<n:
        a.append((s[i],i))
        i+=1
    a.sort()
    i=0
    while i<n:
        p[i]=a[i][1]
        i+=1
    c[p[0]]=0
    i=1
    while i<n:
        c[p[i]]=c[p[i-1]]+(a[i][0]!=a[i-1][0])
        i+=1
    k=0
    while (1<<k)<n:
        pp=[]
        i=0
        while i<n:
            pp.append((p[i]-(1<<k))%n)
            i+=1
        pp.sort(key=lambda x:(c[x], c[(x+(1<<k))%n]))
        p=pp
        cc=[0]*n
        cc[p[0]]=0
        i=1
        while i<n:
            pr=(c[p[i-1]], c[(p[i-1]+(1<<k))%n])
            nw=(c[p[i]], c[(p[i]+(1<<k))%n])
            cc[p[i]]=cc[p[i-1]]+(pr!=nw)
            i+=1
        c=cc
        k+=1
    return p[1:]

def solve3():
    s=input().strip()
    n=len(s)
    sa=sufarr(s)
    pos=0
    i=0
    while i<n:
        if sa[i]==0:
            pos=i+1
            break
        i+=1
    print(pos)
    out=[]
    i=0
    while i<n:
        st=sa[i]
        out.append(s[(st-1)%n])
        i+=1
    print(''.join(out))

solve3()
```
